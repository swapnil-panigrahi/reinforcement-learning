## Policy Evaluation Gridworld

```python
import gym
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = np.zeros((4, 4))

transitions = {}
policy = {}

for i in range(4):
    for j in range(4):
        transitions[(i, j)] = {}
        
        transitions[(i, j)]['N'] = ((i, j-1), -1) if j!=0 else ((i, j), -1)
        transitions[(i, j)]['E'] = ((i, j+1), -1) if j!=3 else ((i, j), -1)
        transitions[(i, j)]['W'] = ((i-1, j), -1) if i!=0 else ((i, j), -1)
        transitions[(i, j)]['S'] = ((i+1, j), -1) if i!=3 else ((i, j), -1)
        
transitions[(0, 0)]['N'] = ((0, 0), 0)
transitions[(0, 0)]['E'] = ((0, 0), 0)
transitions[(0, 0)]['W'] = ((0, 0), 0)
transitions[(0, 0)]['S'] = ((0, 0), 0)

transitions[(3, 3)]['N'] = ((3, 3), 0)
transitions[(3, 3)]['E'] = ((3, 3), 0)
transitions[(3, 3)]['W'] = ((3, 3), 0)
transitions[(3, 3)]['S'] = ((3, 3), 0)

gamma = 1
theta = 0.01

for state in transitions:
    policy[state] = ['N', 'E', 'W', 'S']

stable = False
while not stable:
    v = np.zeros((4, 4))
    while True:
        delta = 0
        vOld = v.copy()
        print(v)
        for state in transitions:
            v[state] = np.sum(
                1 / len(policy[state]) * (reward + gamma * vOld[nextState])
                for action in policy[state]
                for nextState, reward in [transitions[state][action]]
                )
            delta = max(delta, abs(vOld[state] - v[state]))
        if delta < theta:
            break
    
    print(policy)
    stable = True
    for state in transitions:
        bestAction = []
        bestValue = float('-inf')
        
        for action in transitions[state]:
            nextState, reward = transitions[state][action]
            actionValue = reward + gamma * v[nextState]
            
            if actionValue > bestValue:
                bestAction = [action]
                bestValue = actionValue
            elif actionValue == bestValue:
                bestAction.append(action)
        
        policy[state] = sorted(policy[state])
        bestAction = sorted(bestAction)
        
        if policy[state] != bestAction:
            stable = False
            policy[state] = bestAction
    
print("Optimal State Value:")
print(v)
print("Optimal Policy:")
print(policy)
```
